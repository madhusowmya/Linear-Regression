```python
hr_test = df_test['hr'].values.tolist()
len(hr_test)
```

```python
mnth_test = df_test['mnth'].values.tolist()
len(mnth_test)
```

```python
temp_test = df_test['temp'].values.tolist()
len(temp_test)
```

```python
wd_test = df_test['weekday'].values.tolist()
len(wd_test)
```

```python
hum_test = df_test['hum'].values.tolist()
len(hum_test)
```

```python
ws_test = df_test['windspeed'].values.tolist()
len(ws_test)
```

```python
casual_test = df_test['casual'].values.tolist()
len(ws_test)
```

```python
registered_test = df_test['registered'].values.tolist()
len(registered_test)
```

```python
#using 6 attributes
pred = []
for i in range(len(casual_test)):
    P = (a1*mnth_test[i])+(a2*hr_test[i])+(a3*wd_test[i])+(a4*temp_test[i])+(a5*hum_test[i])+(a6*ws_test[i])
    pred.append(P)
summ=0
for i in range(len(casual_test)):
    summ=summ+((casual_test[i]-pred[i])**2)
    
mse=summ/len(casual_test)
print('MSE of casual is:',mse)

predreg = []
for i in range(len(registered_test)):
    PR = (a1r*mnth_test[i])+(a2r*hr_test[i])+(a3r*wd_test[i])+(a4r*temp_test[i])+(a5r*hum_test[i])+(a6r*ws_test[i])
    predreg.append(PR)
summ=0
for i in range(len(registered_test)):
    summ=summ+((registered_test[i]-predreg[i])**2)
    
rmse=summ/len(registered_test)
print('MSE of registered is  is:',rmse)
```

```python
plt.plot(casual_test,pred,'.')
plt.plot(registered_test,predreg,'.')
plt.legend(['casual','registered'])
plt.xlabel("casual(real),registered(real)")
plt.ylabel("casual(predicted),registered(predicted)")
plt.title("Bikesharing prediction for casual and registered using six attributes.\n")

plt.show()
```
